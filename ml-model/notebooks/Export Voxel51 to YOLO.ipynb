# Export from Voxel51 to YOLO Format

```python
export_name="train_split_by_aircraft_model_fixed" # The name for the dataset export
label_field = "multi_class_detections"  # The field from the V51 Samples around which will be used for the Labels for training.
tag = "multi_class_train"
dataset_name = "plane-dataset" # The name of the V51 dataset that will be used
```

```python
export_dir = "/tf/efs/" + export_name + "/"
```

```python
import fiftyone as fo
import math
dataset = fo.load_dataset(dataset_name)
```

## Export the Training and Val Dataset from Voxel 51
Here are some basic stats on the Voxel51 dataset you are going to build training the model on. 
An example of the samples is also printed out. In the Sample, make sure the *label_field* you selected has some detections in it.

```python
from fiftyone import ViewField as F
print("\t\tDataset\n-----------------------------------")
#view = dataset.exists(label_field).match(F("{}.detections".format(label_field)).length()>0).shuffle(seed=51)
#view = dataset.exists(label_field).shuffle(seed=51) # You can add additional things to the query to further refine it. eg .match_tags("good_box")
view = dataset.match_tags(tag).shuffle(seed=51) # You can add additional things to the query to further refine it. eg .match_tags("good_box")
print(view)
print("\n\n\tExample Sample\n-----------------------------------")
print(view.first())
```

```python
# The type of dataset to export
# Any subclass of `fiftyone.types.Dataset` is supported
dataset_type = fo.types.YOLOv4Dataset  # for example

# Export the dataset!
view.export(
    export_dir=export_dir, dataset_type=dataset_type, label_field=label_field
)
```
