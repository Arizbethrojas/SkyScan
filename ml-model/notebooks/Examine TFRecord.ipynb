# Examine TFRecord
Lets you examine an example from a TFRecord

```python
import tensorflow as tf

import numpy as np
import IPython.display as display
```

```python
training_name="planeBox"
train_record_fname = "/tf/dataset-export/" + training_name + "/train/tf.records"
val_record_fname = "/tf/dataset-export/" + training_name + "/val/tf.records"
pet_record = "/tf/dataset-export/pet/pet_faces_train.record-00001-of-00010"
plane_record="/tf/dataset-export/plane-coco/coco_train.record-00000-of-00100"
```

```python
raw_dataset = tf.data.TFRecordDataset(pet_record)

for raw_record in raw_dataset.take(1):
    example = tf.train.Example()
    example.ParseFromString(raw_record.numpy())
    print(example)
```
