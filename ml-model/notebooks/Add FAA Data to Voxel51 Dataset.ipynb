# Add FAA Data to a Voxel 51 Dataset
Adds information about each aircraft from the FAA dataset from: https://opensky-network.org/aircraft-database
It is retrieved based on the ICAO24 field for each sample. Not all aircraft will have an entry.

adapted from: https://voxel51.com/docs/fiftyone/recipes/model_inference.html

```python
DATASET_NAME="jsm-test-dataset"
```

```python
import os

import fiftyone as fo
import pandas as pd
```

```python
# load data
planes = pd.read_csv("./aircraftDatabase.csv", index_col='icao24')
dataset = fo.load_dataset(DATASET_NAME)
```

```python
def add_plane_data(sample):
    """Add FAA data to voxel51 database.
    
    Add model and manufacturer name to row.
    
    Args:
        sample - a voxel51 row
    
    Returns:
        null
    """

    # render plane_id in lowercase letters
    plane_id = sample["icao24"].label.lower()

    try:
        plane = planes.loc[plane_id.lower()]
        # Check for valid row with all columns present
        if plane.size == 26:

            #print("Adding metadata for plane {} onto image {}".format(plane["icao24"].values[0], image["external_id"]))
            if isinstance(plane["model"], str):
                sample["model"] = fo.Classification(label=plane["model"] )
            if isinstance(plane["manufacturername"], str):
                sample["manufacturer"] = fo.Classification(label=plane["manufacturername"])
            if isinstance(plane["operatorcallsign"], str):
                sample["operatorcallsign"] = fo.Classification(label=plane["operatorcallsign"])

            sample.save()
        else:
            print(plane.size)
    except KeyError:
        print("FAA Data entry not found for: {}".format(plane_id))        
```

## Add FAA data to each dataset row

```python
for sample in dataset:
    add_plane_data(sample)
```

### Normalize the Model
